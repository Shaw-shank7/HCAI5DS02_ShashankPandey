```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
import numpy as np

# 1. Sports Performance

# Player sprint times (used as heights in this context)
player_heights = [5.1, 4.8, 5.0, 4.9, 5.3, 4.7, 4.9, 4.8, 5.2, 5.0]

# Convert to NumPy array for computation
heights = np.array(player_heights)

# Calculate mean
mean_height = np.mean(heights)

# Calculate sample standard deviation (ddof=1 for sample)
std_height = np.std(heights, ddof=1)

# Calculate Coefficient of Variation (CV)
cv_height = (std_height / mean_height) * 100

# Print results with proper formatting
print("=== Sports Performance Summary ===")
print(f"Mean Height: {mean_height:.2f} ft")
print(f"Standard Deviation: {std_height:.2f} ft")
print(f"Coefficient of Variation (CV): {cv_height:.2f}%")

# Interpret the spread
if cv_height < 10:
    print("Interpretation: The heights are tightly clustered.")
else:
    print("Interpretation: The heights are highly variable.")
```

 Are the player’s times tightly clustered or highly variable?

**Answer:**
The coefficient for variation is 3.8 which means the player's times are little seperated so this is tightly clustered.


```python
import numpy as np

# Original sprint times with an outlier (3.8 seconds instead of ~5)
heights_with_outlier = [5.1, 4.8, 5.0, 4.9, 5.3, 4.7, 4.9, 4.8, 5.2, 3.8]
data_with_outlier = np.array(heights_with_outlier)

# Recalculate statistics
mean_outlier = np.mean(data_with_outlier)
std_outlier = np.std(data_with_outlier, ddof=1)
cv_outlier = (std_outlier / mean_outlier) * 100

# Print refined results
print("=== With Outlier (3.8 seconds) ===")
print(f"New Mean: {mean_outlier:.2f} ft")
print(f"New Standard Deviation: {std_outlier:.2f} ft")
print(f"New Coefficient of Variation: {cv_outlier:.2f}%")

# Explanation of impact
print("\nImpact of the Outlier:")
print("- The mean has decreased.")
print("- The standard deviation has increased.")
print("- The Coefficient of Variation is now higher.")
print("- This means the data is less tightly clustered and more variable.")
print("- Using the median instead of the mean would reduce the impact of such outliers.")
```

### Recommend whether mean or median should be used for performance reporting.

The median should be used for performance reporting because it is not affected by outliers. In this case, a single mistake (3.8 seconds) lowered the mean and increased variability. The median gives a more accurate and fair representation of typical sprint times, especially in small datasets.

2. Patient Blood Pressure Reading

```python
import numpy as np

bp = [118, 122, 125, 130, 135, 138, 142, 144, 146, 150, 152, 155, 160, 162, 165]

mean_bp = np.mean(bp)
median_bp = np.median(bp)
std_bp = np.std(bp, ddof=1)

print(f"Mean: {mean_bp: .2f}")
print(f"Median: {median_bp}")
print(f"Standard Deviation: {std_bp: .2f}")
```

```python
import matplotlib.pyplot as plt

plt.hist(bp, bins=6, color='#FF7F50', edgecolor='#2E2E2E', alpha=0.8)  # Coral bars with dark edges
plt.title("Systolic Blood Pressure Distribution", fontsize=14, fontweight='bold', color='#4B0082')
plt.xlabel("Blood Pressure (mmHg)", fontsize=12, color='#4B0082')
plt.ylabel("Count", fontsize=12, color='#4B0082')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Observation:
# The histogram shows a slight right skew with values stretching toward the higher BP range.
```

```python
import numpy as np

q1 = np.percentile(bp, 25)
q3 = np.percentile(bp, 75)
iqr = q3 - q1

print(f"IQR (Interquartile Range): {iqr}")

# Calculate fences to detect outliers
lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr

print(f"Lower Fence: {lower_fence}")
print(f"Upper Fence: {upper_fence}")

# Identify any unusually high blood pressure values (outliers)
outliers = [x for x in bp if x > upper_fence or x < lower_fence]

print(f"Outliers: {outliers if outliers else 'No outliers detected'}")
```

### Explain whether these statistics support that the clinic population has a normal range of BP levels.

The mean and median are close, showing the data is fairly symmetric. The histogram’s slight right skew and any outliers indicate a few patients have higher blood pressure. Overall, most patients have normal BP levels, but some may have elevated readings.


3. Retails Sales Summary

```python
import numpy as np
from scipy import stats

sales = [212, 198, 245, 210, 230, 185, 270, 205, 190, 250, 260, 225, 215, 195]

mean_sales = np.mean(sales)
median_sales = np.median(sales)
mode_sales = stats.mode(sales, keepdims=False).mode
range_sales = np.ptp(sales)
var_sales = np.var(sales, ddof=1)
sd_sales = np.std(sales, ddof=1)

print(f"Mean: {mean_sales: .4f}")
print(f"Median: {median_sales}")
print(f"Mode: {mode_sales}")
print(f"Range: {range_sales}")
print(f"Variance: {var_sales: .4f}")
print(f"Std Dev: {sd_sales: .4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

days = range(1, len(sales) + 1)

plt.bar(days, sales, color='#FFA500')  # orange color
plt.title("Daily Sales", fontsize=14, fontweight='bold')
plt.xlabel("Day", fontsize=12)
plt.ylabel("Sales", fontsize=12)

# Plot mean line
plt.axhline(mean_sales, color='red', linestyle='--', label=f"Mean = {mean_sales:.2f}")

# Highlight max and min sales bars
max_sale = max(sales)
min_sale = min(sales)
max_day = sales.index(max_sale) + 1
min_day = sales.index(min_sale) + 1

plt.bar(max_day, max_sale, color='green')
plt.bar(min_day, min_sale, color='blue')

# Annotate max and min sales
plt.text(max_day, max_sale + 5, f"High: {max_sale}", ha='center', color='green', fontweight='bold')
plt.text(min_day, min_sale + 5, f"Low: {min_sale}", ha='center', color='blue', fontweight='bold')

plt.legend()
plt.tight_layout()
plt.show()
```

### Comment on consistency of sales - do the spread and measures indicate a steady flow?

The sales data shows moderate variability. While the mean and median are close, indicating some consistency, the range and standard deviation suggest noticeable fluctuations. Therefore, sales are somewhat steady but not completely consistent day-to-day.

### Suppose Sunday sales are usually 20% lower than the other days. How would this affect interpretation?

If Sunday sales are typically 20% lower, this could pull down the mean and increase the variability, making the data seem less consistent than it actually is on regular days. It suggests the dip is predictable, not random, so Sunday should be treated separately in analysis to avoid misleading conclusions.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated data for returning and new users
np.random.seed(42)
returning_users = np.random.normal(120, 15, 30)  # Returning users: higher mean, lower spread
new_users = np.random.normal(90, 25, 30)         # New users: lower mean, higher spread

# Calculate mean and standard deviation
mean_returning = np.mean(returning_users)
std_returning = np.std(returning_users, ddof=1)
mean_new = np.mean(new_users)
std_new = np.std(new_users, ddof=1)

# Coefficient of Variation (CV)
cv_returning = (std_returning / mean_returning) * 100
cv_new = (std_new / mean_new) * 100

print(f"Returning Users - Mean: ${mean_returning:.2f}, CV: {cv_returning:.2f}%")
print(f"New Users - Mean: ${mean_new:.2f}, CV: {cv_new:.2f}%")

# Interpretation: who is more variable
if cv_returning > cv_new:
    print("Returning users are more variable relative to their mean.")
else:
    print("New users are more variable relative to their mean.")

# Skewness observation:
# → New users' spending is likely more right-skewed (due to occasional large purchases).
# → This pulls the mean up, so it may not reflect a typical spending behavior.

# Campaign effect:
# → If a campaign boosts new user spending for a few days:
#    - Mean increases
#    - Standard deviation may rise
#    - CV and skewness both increase
# → These changes may give a false impression of improvement unless median and IQR are also checked.
# → Median gives a better picture of typical user behavior in such cases.

# Visualization: Boxplot to compare spending distributions
plt.boxplot([returning_users, new_users],
            labels=["Returning Users", "New Users"],
            patch_artist=True,
            boxprops=dict(facecolor="#90EE90"),
            medianprops=dict(color="black"))
plt.ylabel("Cart Value ($)")
plt.title("Cart Value Comparison: Returning vs New Users")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Descriptive techniques for business insights:
# 1. Moving average (e.g., 7-day) could help smooth out random spikes in daily sales and show real trends.
# 2. Boxplots easily reveal central tendency and outliers for each user type.
#    → Useful for spotting potential high spenders or inconsistent patterns.
# 3. (Bonus) Cohort analysis: track user behavior over time to see if new users turn into returning customers or increase spending.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/dataset/daily_sales.csv')

# Check the unique categories (optional)
print(df['category'].unique())  # To confirm category names

# Define palette matching your exact category names
palette = {'Category A': 'blue', 'Category B': 'orange'}

# Boxplot: visualize total sales distribution by category
plt.figure(figsize=(8,6))
sns.boxplot(x='category', y='total_sales', data=df, palette=palette)
plt.title('Boxplot of Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Histogram: distribution of total sales for each category with KDE
plt.figure(figsize=(10,6))
sns.histplot(data=df[df['category'] == 'Category A'], x='total_sales', color='blue', label='Category A', kde=True, alpha=0.6)
sns.histplot(data=df[df['category'] == 'Category B'], x='total_sales', color='orange', label='Category B', kde=True, alpha=0.6)
plt.title('Histogram of Total Sales by Category')
plt.xlabel('Total Sales')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Interpretation:
# - The median (center line in the boxplot) and peaks in the histogram show the central tendency of sales.
# - The width of the box (IQR) and spread of the histogram indicate variability.
# - Wider boxes and broader histograms mean more variation in daily sales for that category.
```

```python
import pandas as pd

cv_data = []

# Use exact category names from your data
for cat in ['Category A', 'Category B']:
    df_cat = df[df['category'] == cat]
    for col in ['total_sales', 'avg_basket']:
        mean_val = df_cat[col].mean()
        std_val = df_cat[col].std()
        cv = (std_val / mean_val) * 100
        cv_data.append((cat, col, cv))

cv_df = pd.DataFrame(cv_data, columns=['Category', 'Metric', 'CV'])
print(cv_df)

# Interpretation:
# The category and metric with higher CV (%) shows greater relative variability.
# For example, if 'Category B' has higher CV in 'total_sales', its sales are less consistent.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Convert 'date' to datetime if not already done
df['date'] = pd.to_datetime(df['date'])

# Sort data by date to ensure correct rolling calculation
df = df.sort_values('date')

# Calculate 7-day moving average of total sales by category
df['total_sales_ma'] = df.groupby('category')['total_sales'].transform(
    lambda x: x.rolling(window=7, min_periods=1).mean()
)

# Plot the 7-day moving average time series
plt.figure(figsize=(12, 6))
sns.lineplot(data=df[df['category'] == 'Category A'], x='date', y='total_sales_ma', label='Category A (7-day MA)')
sns.lineplot(data=df[df['category'] == 'Category B'], x='date', y='total_sales_ma', label='Category B (7-day MA)')
plt.xticks(rotation=45)
plt.title('7-Day Moving Average of Total Sales')
plt.xlabel('Date')
plt.ylabel('Total Sales (7-day MA)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Interpretation:
# - The 7-day moving average smooths out daily fluctuations, highlighting underlying trends.
# - A stable, smooth curve indicates consistent sales, aiding reliable forecasting and inventory planning.
# - Large fluctuations or spikes even after smoothing suggest volatility, requiring cautious business strategies.
# - Businesses can use this smoothed data to anticipate demand cycles and optimize marketing or stock levels.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"/content/drive/MyDrive/dataset/customer_segments.csv")
print(df.columns)  # Check column names to ensure correctness

# Violin + Strip Plot: shows distribution shape, modality, and outliers
plt.figure(figsize=(10, 6))
sns.violinplot(x='segment', y='transaction_amount', data=df, inner=None, palette='Pastel1')
sns.stripplot(x='segment', y='transaction_amount', data=df, color='black', size=3, jitter=0.2)
plt.title('Violin & Strip Plot of Transaction Amount by Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Transaction Amount')
plt.show()

# Interpretation:
# - Violin plots show distribution shape and modality (peaks).
# - Strip plots reveal individual data points and possible outliers.
# - Skewness is visible if the violin plot is asymmetric.
# - Presence of wide spread or outliers can be spotted with strip plot dots far from the bulk.

# KDE Plot: density estimation to visualize spending variability and central tendencies
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df[df['segment'] == 'Student']['transaction_amount'], fill=True, label='Student', alpha=0.5)
sns.kdeplot(data=df[df['segment'] == 'Working Professionals']['transaction_amount'], fill=True, label='Working Professionals', alpha=0.5)
plt.title('Density Plot of Transaction Amount by Segment')
plt.xlabel('Transaction Amount')
plt.ylabel('Density')
plt.legend()
plt.show()

# Interpretation:
# - Wider KDE curves indicate higher variability in transaction amounts.
# - A KDE curve shifted to the right shows higher spending tendencies.
# - Compare peaks and spread to infer which segment tends to spend more and is more consistent.
```

```python
import numpy as np
import pandas as pd

summary_stats = []

# Loop through each customer segment and calculate statistics
for segment in df['segment'].unique():
    data = df[df['segment'] == segment]['transaction_amount']
    mean = data.mean()                     # Average spending
    median = data.median()                 # Middle value spending
    q75, q25 = np.percentile(data, [75, 25])
    iqr = q75 - q25                       # Interquartile range: spread of middle 50%
    cv = (data.std() / mean) * 100        # Coefficient of Variation: relative variability

    summary_stats.append([segment, mean, median, iqr, cv])

# Create a DataFrame to display results neatly
summary_df = pd.DataFrame(summary_stats, columns=['Segment', 'Mean', 'Median', 'IQR', 'CV (%)'])
print(summary_df)
```
